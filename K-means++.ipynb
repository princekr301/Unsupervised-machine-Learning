```python
import pandas as pd
pd.set_option("display.max_rows",None)
import warnings
warnings.filterwarnings("ignore")
```

```python
a="C:/Users/Prince kumar/Downloads/Mall_Customers (1).csv"
```

```python
df=pd.read_csv(a)
```

```python
df
```

```python
x=df.iloc[0:,3:]
x
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wc=[]
for i in range(1,11):
    Kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    Kmeans.fit(x)
    wc.append(Kmeans.inertia_)
    
plt.figure(figsize=(8,6))
plt.plot(range(1,11),wc,marker="o",color="orange")
plt.title("The Elbow Method",fontweight="bold")
plt.xlabel("Number of cluster")
plt.ylabel("wc")
plt.grid()
plt.show()
```

```python
#fitting the k-maenas data set
Kmeans=KMeans(n_clusters=5,init="k-means++",random_state=41)
fits=Kmeans.fit_predict(x)
print(fits)
```

```python
df["clusters"]=fits
```

```python
df
```

```python
a=df[df.clusters==0]
b=df[df.clusters==1]
c=df[df.clusters==2]
d=df[df.clusters==3]
e=df[df.clusters==4]
```

```python
plt.figure(figsize=(9,8))
plt.scatter(a["Annual Income (k$)"],a["Spending Score (1-100)"],s=50,c="red",label="first_Cluster")
plt.scatter(b["Annual Income (k$)"],b["Spending Score (1-100)"],s=50,c="yellow",label="Second_Cluster")
plt.scatter(c["Annual Income (k$)"],c["Spending Score (1-100)"],s=50,c="blue",label="third_Cluster")
plt.scatter(d["Annual Income (k$)"],d["Spending Score (1-100)"],s=50,c="orange",label="forth_Cluster")
plt.scatter(e["Annual Income (k$)"],e["Spending Score (1-100)"],s=50,c="green",label="fifth_Cluster")
plt.scatter(Kmeans.cluster_centers_[:,0],Kmeans.cluster_centers_[:,1],s=300,c="black",label="Centroid")
plt.legend(loc="best")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Clusters of costmber",fontweight="bold",color="darkgreen")
plt.show
```

# Date - 23-01-2023          [Kmeans unsupervised machine learning]

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

```python
a="C:/Users/Prince kumar/Downloads/Mall_Customers (1).csv"
df=pd.read_csv(a)
df
```

```python
x=df[["Age","Annual Income (k$)"]].values
```

```python
y=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init="k-means++",random_state=42)
    km.fit(x)
    y.append(km.inertia_)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,11),y,marker="o",color="orange")
plt.xlabel("Number of cluster")
plt.ylabel("Number of range")
plt.title("Extract the number of cluster",fontweight="bold")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=42)
fits=kmeans.fit_predict(x)
fits
```

```python
lab=kmeans.labels_
```

```python
kmeans.inertia_
```

```python
cl=kmeans.cluster_centers_
cl
```

```python
df["group"]=lab
```

```python
df.head()
```

```python
a=df[df.group==0]
b=df[df.group==1]
c=df[df.group==2]
d=df[df.group==3]
e=df[df.group==4]
#f=df[df.group==5]
```

```python
#plt.figure(figsize=(14,10))
fig=plt.figure(figsize=(8,8))
ax=fig.add_axes([0,0,.95,1],projection="3d",elev=35,azim=170)
plt.scatter(a["Age"],a["Annual Income (k$)"],s=50,c="red",label="group-1")
plt.scatter(b["Age"],b["Annual Income (k$)"],s=50,c="green",label="group-2")
plt.scatter(c["Age"],c["Annual Income (k$)"],s=50,c="orange",label="group-3")
plt.scatter(d["Age"],d["Annual Income (k$)"],s=50,c="yellow",label="group-4")
plt.scatter(e["Age"],e["Annual Income (k$)"],s=50,c="blue",label="group-5")
plt.scatter(cl[:,0],cl[:,1],s=150,c="black",label="Centroid")
plt.legend(loc=True)
plt.xlabel("Number of cluster")
plt.ylabel("range of cluster")
plt.title("K-Means++ Algorithms in Unsupervised Machine Learning",fontweight="bold",color="blue")
plt.show()
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_axes([0.5,0.5,.90,1],projection="3d",elev=45,azim=160)
plt.scatter(cl[:,0],cl[:,1],s=150)
plt.show()
```

```python
from sklearn.metrics import homogeneity_score,silhouette_score
```

```python
ac=silhouette_score(x,kmeans.labels_)
```

```python
print("Accuracy of k-means Algorithms :- ",ac)
```
